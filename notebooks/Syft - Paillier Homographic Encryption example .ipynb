#  Paillier Homographic Encryption example 

This is a warm up code of what the possibilites are in the present library. It gives a snippet of the vision of the OpenMined project

```python
import warnings
import numpy as np
import phe as paillier
from syft.he.paillier.keys import KeyPair
from syft.nn.linear import LinearClassifier
import numpy as np
from sklearn.datasets import load_diabetes
```

```python
pubkey,prikey = KeyPair().generate(n_length=2048)
```

```python
diabetes = load_diabetes()
target = diabetes.target
input = diabetes.data
```

```python
model = LinearClassifier(desc="DiabetesClassifier",n_inputs=10,n_labels=1).encrypt(pubkey)
```

```python
for iter in range(3):
    for i in range(len(input)):
        print("Grads:" + str((model.learn(input=input[i],target=target[i],alpha=0.5))))
```

```python
model = model.decrypt(prikey)
```

```python
for i in range(len(input)):
    print(model.forward(input[i]))
```
